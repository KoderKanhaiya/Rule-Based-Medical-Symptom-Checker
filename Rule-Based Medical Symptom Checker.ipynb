```python
medical_knowledge = {
    "Flu": ["fever", "cough", "sore throat", "body pain"],
    "Common Cold": ["cough", "sneezing", "runny nose"],
    "Malaria": ["fever", "chills", "sweating", "headache"],
    "Food Poisoning": ["vomiting", "diarrhea", "stomach pain"],
    "Heart Risk": ["chest pain", "high blood pressure", "fatigue"]
}


def show_symptoms():
    print("\nAvailable symptoms you can enter:\n")

    all_symptoms = set()
    for symptom_list in medical_knowledge.values():
        for symptom in symptom_list:
            all_symptoms.add(symptom)

    for symptom in sorted(all_symptoms):
        print("-", symptom)


def get_user_input():
    show_symptoms()
    user_input = input("\nEnter your symptoms separated by commas: ")

    user_symptoms = user_input.split(",")
    cleaned_symptoms = []

    for symptom in user_symptoms:
        cleaned_symptoms.append(symptom.strip().lower())

    return cleaned_symptoms


def diagnose(symptoms):
    results = {}

    for disease, disease_symptoms in medical_knowledge.items():
        match_count = 0

        for symptom in disease_symptoms:
            if symptom in symptoms:
                match_count += 1

        percentage = (match_count / len(disease_symptoms)) * 100
        results[disease] = percentage

    return results


def show_results(results):
    print("\n=Diagnosis Report =\n")

    sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)

    for disease, percent in sorted_results:
        print(f"{disease}: {percent:.2f}% match")

    best_disease, best_score = sorted_results[0]

    if best_score > 0:
        print("\nMost probable condition:", best_disease)
    else:
        print("\nNo matching disease found.")


def main():
    print("======================================")
    print("      Medical Symptom Checker")
    print("======================================")

    user_symptoms = get_user_input()
    diagnosis = diagnose(user_symptoms)
    show_results(diagnosis)


main()
```
